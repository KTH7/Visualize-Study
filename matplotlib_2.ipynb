```python
import pandas as pd
df = pd.read_csv('./dataset/1__2.csv',encoding='euc-kr',index_col='')
df.head()
df.plot()
fig=plt.figure(figsize=(20,10))
ax=fig.add_subplot(111)
wt=np.array(range(len(df)))
w=0.1

for i in df.columns:
    ax.bar(wt,df[i],width=w,label=i)
    wt=wt+w

ax.set_xticks(np.array(range(len(df))))
ax.set_xticklabels(df.index)
ax.set_ylabel(' ')
ax.legend()
plt.show()
```

```python
s_df=pd.read_csv('./dataset/.csv',encoding='euc-kr',index_col=0)
s_df.head(10)

fig=plt.figure(figsize=(10,5))
s_ax_1=fig.add_subplot(111)
plt.style.use('ggplot')

s_ax_1.plot(s_df.index,s_df[''],color='g',label='')

s_ax_2=s_ax_1.twinx()
s_ax_2.bar(s_df.index,s_df[''],color='r',label='',width=2)

s_ax_1.set_yticks([i*2000 for i in range(5)])
s_ax_1.set_ylabel('')
s_ax_1.set_xlabel('')

s_ax_2.set_yticks([i*50000 for i in range(5)])
s_ax_2.set_ylabel('')

s_ax_1.set_title(' ')

s_ax_1.legend(loc=1)
s_ax_2.legend(loc=2)

plt.show()
```

```python
import pandas as pd
df=pd.read_csv('./dataset/ani_info.csv',encoding='euc-kr',index_col='anime_id')
df.head(10)

df.loc[df['vote_no']>=800000,['name','vote_no']]

df.loc[(df['vote_no']>=600000)&(df['rating']>=8),['name','rating']]

#
types=df['type'].unique()
types

fig=plt.figure(figsize=(15,7))
ax=fig.add_subplot(111)

for t in types:
    x=df.loc[df['type']==t,'vote_no']
    y=df.loc[df['type']==t,'rating']
    
    ax.scatter(x,y,alpha=0.5,label=t)# alpha: 
    
ax.set_title('   ')
ax.set_xlabel('  ')
ax.set_ylabel('')

ax.legend(loc='lower right',fontsize=10)
plt.show()
```

```python
import pandas as pd
c_df=pd.read_csv('./dataset/1__2.csv',encoding='euc-kr',index_col='')

c_df.head(10)

c_df_2004=c_df.iloc[2]
c_df_2005=c_df.iloc[3]

fig=plt.figure(figsize=(10,5))
ax1=fig.add_subplot(121)
ax2=fig.add_subplot(122)

color=('red','blue','yellow','green')

ax1.pie(c_df_2004,
        explode=(0,0.1,0,0),
        labels=c_df_2004.index,
        autopct='%1.0f%%',
        colors=color,
        startangle=90,counterclock=False
        )

ax2.pie(c_df_2005,
        explode=(0,0.1,0,0),
        labels=c_df_2005.index,
        autopct='%1.0f%%',
        colors=color,
        startangle=90,counterclock=False
        )

ax1.set_title('2004  ')
ax2.set_title('2005  ')

fig.subplots_adjust(wspace=0.2)
plt.show()
```
